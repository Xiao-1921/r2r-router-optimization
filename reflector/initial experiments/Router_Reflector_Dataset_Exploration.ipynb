### Explorer Helper


```python
import pandas as pd
import random
from datasets import load_dataset, get_dataset_split_names

def explore(path, name=None, split="train", n=3, sample=250, seed=42):
    print(f"{'='*60}")
    print(f"  {path}  |  config={name}")
    print(f"{'='*60}")

    # ── Available splits ──────────────────────────────────────────
    try:
        splits = get_dataset_split_names(path, name) if name else get_dataset_split_names(path)
        print(f"  Available splits : {splits}")
    except Exception as e:
        print(f"  Could not fetch split names: {e}")

    # ── Load ──────────────────────────────────────────────────────
    print(f"  Loading split    : '{split}' ...")
    try:
        ds = load_dataset(path, name, split=split) if name else load_dataset(path, split=split)
    except Exception as e:
        print(f"  ERROR loading: {e}")
        return None

    df_full = ds.to_pandas()
    print(f"\n  ── Full dataset stats {'─'*38}")
    print(f"  Total rows  : {len(df_full)}")
    print(f"  Columns     : {df_full.columns.tolist()}")
    print()

    # ── Per-column stats ──────────────────────────────────────────
    for col in df_full.columns:
      print(f"  [{col}]")

      # Type
      print(f"    dtype      : {df_full[col].dtype}")

      # Sample value
      print(f"    sample val : {repr(df_full[col].iloc[0])[:120]}")

      # Null count
      print(f"    nulls      : {df_full[col].isna().sum()}")

      # Unique values or value counts — try, catch if unhashable
      try:
          n_unique = df_full[col].nunique()
          print(f"    n_unique   : {n_unique}")
          if n_unique <= 20:
              print(f"    counts     : {df_full[col].value_counts().to_dict()}")
      except TypeError:
          print(f"    n_unique   : unhashable (list/array column)")

      # Length distribution if iterable
      try:
          lengths = df_full[col].apply(len)
          print(f"    len dist   : {lengths.value_counts().to_dict()}")
      except (TypeError, AttributeError):
          pass

      print()

    # ── Pretty-print n examples ───────────────────────────────────
    print(f"  ── First {n} examples {'─'*38}")
    for i in range(min(n, len(df_full))):
        print(f"  ── Example {i} {'─'*40}")
        for k, v in ds[i].items():
            print(f"  {k:20s}: {repr(v)[:250]}")
        print()

    # ── Sample into DataFrame ─────────────────────────────────────
    random.seed(seed)
    if sample is None:
        df_sampled = df_full.reset_index(drop=True)
    else:
        indices = random.sample(range(len(df_full)), min(sample, len(df_full)))
        df_sampled = df_full.iloc[indices].reset_index(drop=True)
    print(f"  Returning sampled DataFrame — {len(df_sampled)} rows (seed={seed})")
    return df_sampled
```

## MMLU
Link: https://huggingface.co/datasets/cais/mmlu



```python
df_mmlu = explore("cais/mmlu", name="all", split="auxiliary_train", n=3, sample=250)
```

```python
print(df_mmlu["subject"].value_counts())
print()

# Check other splits
for split in ["test", "validation", "dev"]:
    ds = load_dataset("cais/mmlu", "all", split=split)
    print(f"{split}: {len(ds)} rows, sample subjects: {list(set(ds['subject'][:20]))[:5]}")
```

```python
# Number of questions per option count
df_mmlu["choices"].apply(len).value_counts()
```

```python
from datasets import get_dataset_config_names
import pandas as pd

# ── Get all subjects ─────────────────────────────────────────────
subjects = get_dataset_config_names("cais/mmlu")

dfs = []

# ── Load full data per subject ───────────────────────────────────
for subj in subjects:
    print(f"\nLoading subject: {subj}")

    df = explore(
        "cais/mmlu",
        name=subj,
        split="auxiliary_train",
        sample=None   # effectively disables sampling
    )

    if df is not None:
        df["subject"] = subj
        dfs.append(df)

# ── Combine ─────────────────────────────────────────────────────
df_mmlu_all = pd.concat(dfs, ignore_index=True)

print("\n" + "="*60)
print("COMBINED DATASET OVERVIEW")
print("="*60)

print(f"Total rows           : {len(df_mmlu_all)}")
print(f"Columns              : {df_mmlu_all.columns.tolist()}")

# ── Column-wise quick stats ──────────────────────────────────────
print("\nColumn summary:")
for col in df_mmlu_all.columns:
    print(f"\n[{col}]")
    print(f"  dtype      : {df_mmlu_all[col].dtype}")
    print(f"  nulls      : {df_mmlu_all[col].isna().sum()}")

    try:
        n_unique = df_mmlu_all[col].nunique()
        print(f"  n_unique   : {n_unique}")
        if n_unique <= 20:
            print(f"  counts     : {df_mmlu_all[col].value_counts().to_dict()}")
    except TypeError:
        print("  n_unique   : unhashable")

# ── Per-subject distribution ─────────────────────────────────────
print("\n" + "="*60)
print("QUESTIONS PER SUBJECT")
print("="*60)

subject_counts = (
    df_mmlu_all["subject"]
    .value_counts()
    .sort_index()
)

print(subject_counts)

print("\nSummary stats:")
print(f"Total subjects       : {subject_counts.shape[0]}")
print(f"Min questions        : {subject_counts.min()}")
print(f"Max questions        : {subject_counts.max()}")
print(f"Mean questions       : {subject_counts.mean():.2f}")

# ── Optional: inspect imbalance ──────────────────────────────────
print("\nTop 5 largest subjects:")
print(subject_counts.sort_values(ascending=False).head())

print("\nTop 5 smallest subjects:")
print(subject_counts.sort_values().head())
```
